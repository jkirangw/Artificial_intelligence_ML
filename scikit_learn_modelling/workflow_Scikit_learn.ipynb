```python
#lets teach computers by building models for disease prediction, diagnosis, prevention and treatment
#normally we input data and provide a function or algorithm to process the input into a desired output (programming)
#However with machine learning we give the computer input and expected output and it will learn and create a model that generates the desired output.
#so instead of us writing the algorihm/function the machine writes the function (model/algorithm/brain(gets input and generates output desired): model()
#conda activate env
#jupyter notebook loaded
import pandas as pd
import numpy as np
```

```python
#Get data ready
#Import data (rows/axix=0 and columns/axis=1)
heart_disease_df = pd.read_csv("/home/jkirangw/PROJECT_FOLDER_DATASCIENCE_MACHINELEARNING/Data/heart-disease.csv")
heart_disease_df.head()
```

```python
#We are going to predict target column: heart disease (1) or not (0)
heart_disease_df.columns
```

## Getting data ready by creating feature matrix (x) and labels (y)

```python
# x: features matrix from features columns 
# y: target column (what you are predicting)
#create x features matrix 
x = heart_disease_df.drop("target", axis = 1)

#create y (labels)
y = heart_disease_df["target"]
```

```python
#Choose an estimator or model for the problem (here classification model is needed to classify heart diease or not
#Choose the model and hyperparameters
#RandomForestClassifier is one of the classification models in sklearn. its capable of learning patterns in the data and perform classification
import sklearn
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

#keep the default hyperparameters
clf.get_params()
```

```python
# Now fit the model to training data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
clf.fit(x_train, y_train)
```

```python
#Make prediction
y_preds = clf.predict(x_test)
```

```python
y_preds
```

```python
y_test
```

```python
#Evaluate the model on how it has done on both train data and test data
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds))
```

```python
 confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

```python
#improve the model by trying different amount of estimaters
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying a model with {i} estimators..")
    clf=RandomForestClassifier(n_estimators=i).fit(x_train, y_train)
    print(f"Model accuracy on test set: {clf.score(x_test, y_test) * 100:.2f}")
    print("")
    
```

```python
#Save a model and load it
import pickle
pickle.dump(clf, open("random_forest_model_1.pk1", "wb"))
```

```python
loaded_model = pickle.load(open("random_forest_model_1.pk1", "rb"))
```
